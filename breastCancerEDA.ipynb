```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets

sns.set_style('darkgrid')
```

```python
data = datasets.load_breast_cancer()
```

```python
df =pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
df.head()
```

```python
df['diagnose']=data['target']
df['diagnose_target']=df['diagnose'].apply(lambda x: data['target_names'][x])
```

```python
df.head()
```

```python
# shape of dataframe
df.shape
```

```python
# Columns present in dataset
df.info()
```

```python
# Check for null values
df.isnull().sum()
```

```python
df.describe()
```

### **Check correlation between target values and features**

```python
correlation = df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(correlation,annot=True)
```

```python
df.corr()['diagnose'][:].sort_values()[:10]
```

### **Selecting only highly correlated values**

```python
df_corr = df[[
    "worst concave points",
    "worst perimeter",
    "mean concave points",
    "worst radius",
    "mean perimeter",
    "worst area",
    "mean radius",
    "mean area",
    "mean concavity",
    "worst concavity",
    "diagnose",
    "diagnose_target"
]]
```

```python
df_corr.head()
```

```python
df_corr.describe()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df_corr.corr(),annot=True)
```

```python
sns.pairplot(data=df_corr, hue='diagnose_target',diag_kind='hist')
```

### **Understanding from above plots:**
 - Lower values of features signifies benign and higher values signifies malignent

### **Understanding highly related features**

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='diagnose_target',y='worst concave points',data=df_corr)
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(x='diagnose_target',y='worst concave points',data=df_corr)
```

```python
plt.figure(figsize=(8,8))
sns.barplot(y='worst concave points',x ="diagnose_target",data=df_corr)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='worst concave points',y='worst perimeter',hue='diagnose_target',data=df_corr)
```
